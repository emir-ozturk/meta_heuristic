```python
from mealpy import SMA, GA, PSO, ABC, SOS
from kmeans.kmeans_problem import KMeansMetaHeuristicProblem
from kmeans.kmeans_problem_bounds import problem_bounds, termination
from utils.visualize import plot_fitness
from dataset_utils.data import load_diabetes_dataset, load_titanic_dataset, load_cinnamon_dataset
import pandas as pd
```

```python
diabetes_df = pd.read_csv("datasets/diabetes.csv")
attributes = diabetes_df.drop("Outcome", axis=1)
target = diabetes_df["Outcome"]
```

```python
problem = KMeansMetaHeuristicProblem(attributes, target, problem_bounds)
model = PSO.OriginalPSO(epoch=80)
model.solve(problem, termination=termination, seed=10)
```

```python
plot_fitness(model)
```

```python
data = load_titanic_dataset()
problem = KMeansMetaHeuristicProblem(data[0], data[1], problem_bounds)
model = GA.EliteSingleGA(epoch=80)
model.solve(problem, termination=termination, seed=10)
```

```python
plot_fitness(model)
```
