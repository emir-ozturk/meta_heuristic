```python
import pandas as pd

data = pd.read_csv("personality_synthetic_dataset.csv")
```

```python
data.personality_type.value_counts()
```

```python
data.describe().T
```

```python
target_key = {
    "Introvert": 0,
    "Extrovert": 1,
    "Ambivert": 2
}

data["personality_type"] = data["personality_type"].map(target_key)

data.head()
```

```python
from mealpy import SMA, GA, PSO, ABC, SOS
from knn_problem_for_meta_heuristic import KnnProblemForMetaHeuristic
from meta_heuristic_visualize import MetaHeuristicVisualize
from knn_problem_bounds import problem_bounds, termination

attributes = data.drop("personality_type", axis=1)
target = data["personality_type"]
```

```python
problem = KnnProblemForMetaHeuristic(attributes, target, problem_bounds)
model = SOS.OriginalSOS(epoch=50, pop_size=40)
```

```python
model.solve(problem, termination=termination, seed=10)
```

```python
MetaHeuristicVisualize.plot(model)
```
